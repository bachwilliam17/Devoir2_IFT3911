```python
import matplotlib.pyplot as plt
import numpy as np
from fractions import Fraction
```

```python
my_data = np.loadtxt('abstra.txt', delimiter=',', dtype = str)
```

```python
print (my_data[1])
```

```python
Classes = my_data[1:, 1]
Instabilité = my_data[1:, 2]
Abstraction = my_data[1:, 3]
```

```python
for i in range(len(Instabilité)):
    Instabilité[i] = float(sum(Fraction(s) for s in Instabilité[i].split()))
    Abstraction[i] = float(sum(Fraction(s) for s in Abstraction[i].split()))
    
Insta2 = [float(line) for line in Instabilité]
Abstr2 = [float(line) for line in Abstraction]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(Insta2, Abstr2)
for i, txt in enumerate(Classes):
    ax.annotate(txt, (Insta2[i], Abstr2[i]))
    
    
x = np.linspace(0,1,100)
y = 1-x
plt.plot(x, y, '-b', label='y=2x+1')
plt.title('Graph of IA de notre diagramme')
plt.xlabel('I', color='#000000')
plt.ylabel('A', color='#000000')
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([0,1]) 
plt.figsize=(100,100)
plt.grid()
plt.show()
```
